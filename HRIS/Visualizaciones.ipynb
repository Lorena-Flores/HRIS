#### 1.- Gráfico de barras

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
actividades = ["Infraestructura", "Recursos Humanos", "Desarrollo Tecnológico", "Abastecimiento"]
dimensiones = ["Cultura Digital", "Liderazgo Digital", "Automatización", "Innovación", "Gestión de Datos"]

valores = [
    [3, 2, 3, 4, 3],  # Infraestructura
    [2, 3, 2, 3, 2],  # Recursos Humanos
    [4, 3, 4, 4, 4],  # Desarrollo Tecnológico
    [2, 2, 3, 2, 2],  # Abastecimiento
]

colores = ['#99CCFF', '#3399FF', '#0066CC', '#003366'] 

# Configuración del gráfico
x = np.arange(len(dimensiones))
bar_width = 0.2

fig, ax = plt.subplots(figsize=(10, 6))

for i, actividad in enumerate(actividades):
    ax.bar(x + i * bar_width, valores[i], bar_width, color=colores[i], label=actividad)

# Personalización
# ax.set_xlabel('Dimensiones')
ax.set_ylabel('Nivel de Madurez')
ax.set_title('Nivel de Madurez Digital por Actividad y Dimensión')
ax.set_xticks(x + bar_width * 1.5)
ax.set_xticklabels(dimensiones)

# Modificar ubicación de la leyenda
ax.legend(loc='lower center', bbox_to_anchor=(0.5, -0.2), ncol=2)

# Mostrar gráfico
plt.tight_layout()
plt.show()
```

#### 2.- Gráfico radar (Spider Chart)

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
dimensiones = ["Cultura Digital", "Liderazgo Digital", "Automatización", "Innovación", "Gestión de Datos"]
actividades = ["Infraestructura", "Recursos Humanos", "Desarrollo Tecnológico", "Abastecimiento"]

valores = [
    [3, 2, 3, 4, 3],  # Infraestructura
    [2, 3, 2, 3, 2],  # Recursos Humanos
    [4, 3, 4, 4, 4],  # Desarrollo Tecnológico
    [2, 2, 3, 2, 2],  # Abastecimiento
]

# Normalizar los valores para el gráfico radar (1-4 -> 0-1)
valores_normalizados = [[(v - 1) / 3 for v in actividad] for actividad in valores]

# Configuración del gráfico radar
num_vars = len(dimensiones)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # Cerrar el gráfico

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Añadir cada actividad al radar
colores = ['#99CCFF', '#3399FF', '#0066CC', '#003366']  # Tonos de azul
for i, actividad in enumerate(actividades):
    datos = valores_normalizados[i] + valores_normalizados[i][:1]  # Cerrar el círculo
    ax.fill(angles, datos, color=colores[i], alpha=0.25)
    ax.plot(angles, datos, color=colores[i], label=actividad)

# Añadir etiquetas y leyenda
ax.set_xticks(angles[:-1])
ax.set_xticklabels(dimensiones)
ax.set_yticks([0.25, 0.5, 0.75, 1.0])
ax.set_yticklabels(['Bajo', 'Medio-Bajo', 'Medio', 'Alto'])
ax.set_title("KPIs de Madurez Digital por Actividad", y=1.1)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

#### 3.- Gráfico de líneas

```python
import matplotlib.pyplot as plt

# Datos
dimensiones = ["Cultura Digital", "Liderazgo Digital", "Automatización", "Innovación", "Gestión de Datos"]
actividades = ["Infraestructura", "Recursos Humanos", "Desarrollo Tecnológico", "Abastecimiento"]

valores = [
    [3, 2, 3, 4, 3],  # Infraestructura
    [2, 3, 2, 3, 2],  # Recursos Humanos
    [4, 3, 4, 4, 4],  # Desarrollo Tecnológico
    [2, 2, 3, 2, 2],  # Abastecimiento
]

# Configuración del gráfico de líneas
fig, ax = plt.subplots(figsize=(10, 6))
for i, actividad in enumerate(actividades):
    ax.plot(dimensiones, valores[i], marker='o', label=actividad)

# Personalización
ax.set_title('Tendencias en los KPIs de Madurez Digital')
ax.set_xlabel('Dimensiones')
ax.set_ylabel('Nivel de Madurez Digital')
ax.legend(loc='upper left', bbox_to_anchor=(1, 1)) 
plt.grid(True)

# Mostrar gráfico
plt.show()
```

#### 4.- Gráfico de burbujas (Bubble Chart)

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos simulados para dimensiones
dimensiones = ["Cultura Digital", "Liderazgo Digital", "Automatización", "Innovación", "Gestión de Datos"]
impacto = [30, 20, 40, 35, 25]  # Ejemplo del impacto en cada dimensión
prioridad = [3, 4, 5, 2, 1]  # Prioridad (1-5)
valores = [3, 2, 4, 4, 3]  # Niveles de madurez digital

# Crear gráfico de burbujas
fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(dimensiones, valores, s=np.array(impacto) * 10, alpha=0.6, c=prioridad, cmap='viridis')

# Personalización
ax.set_title('Impacto vs Madurez Digital por Dimensión')
ax.set_xlabel('Dimensiones')
ax.set_ylabel('Nivel de Madurez Digital')
cbar = plt.colorbar(scatter)
cbar.set_label('Prioridad (1-5)')

# Mostrar gráfico
plt.show()
```

#### 5.- Mapa de calor (Heatmap)

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Datos
data = {
    "Cultura Digital": [3, 2, 4, 2],
    "Liderazgo Digital": [2, 3, 3, 2],
    "Automatización": [3, 2, 4, 3],
    "Innovación": [4, 3, 4, 2],
    "Gestión de Datos": [3, 2, 4, 2]
}
actividades = ["Infraestructura", "Recursos Humanos", "Desarrollo Tecnológico", "Abastecimiento"]
df = pd.DataFrame(data, index=actividades)

# Crear mapa de calor
plt.figure(figsize=(8, 6))
sns.heatmap(df, annot=True, cmap="Blues", cbar=True)

# Personalización
plt.title('Mapa de Calor: Madurez Digital por Dimensión y Actividad')
plt.xlabel('Dimensiones')
plt.ylabel('Actividades Secundarias')

# Mostrar gráfico
plt.show()
```
